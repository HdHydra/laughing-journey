# Predicting survival of a passanger in the Titanic tragedy

```python
#Importing Llbaries
import  numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#plots are displayed inline
%matplotlib inline
```

```python
#Reading data (' ' is a must for file name)
data = pd.read_csv('data_cleaned.csv')
```

```python
data.shape
```

```python
data.head()
```

```python
#prints null values in each coulumn

data.isnull().sum()
```

```python
#seperating dependant and independant variables
y = data['Survived']
X = data.drop('Survived', axis = 1)
```

```python
#Importing train_test_split library
from sklearn.model_selection import train_test_split
```

```python
#Creating trainin and validation sets
X_train, X_valid, y_train, y_valid = train_test_split(X, y, random_state = 101, stratify = y, test_size = 0.25)
```

```python
#Distribution in training set
y_train.value_counts(normalize = True)
```

```python
#distribution in validation set
y_valid.value_counts(normalize = True)
```

```python
#Shape of taining set
X_train.shape, y_train.shape
```

```python
#shape of validation set
X_valid.shape, y_train.shape
```

```python
#import required library decision tree classifier
from sklearn.tree import DecisionTreeClassifier
```

```python
#how to import decision tree regressor
from sklearn.tree import DecisionTreeRegressor
```

```python
#creating decision tree function
dt_model = DecisionTreeClassifier(random_state=10)
```

```python
#fitting the model
dt_model.fit(X_train,y_train)
```

```python
#accuracy of a model
dt_model.score(X_train, y_train)
```

```python
dt_model.score(X_valid, y_valid)
```

```python
dt_model.predict(X_valid)
```

```python
dt_model.predict_proba(X_valid)
```

```python
y_pred = dt_model.predict_proba(X_valid)[:,1]
```

```python
new_y = []
for i in range (len(y_pred)):
    if y_pred[i] < 0.6:
        new_y.append(0)
    else:
        new_y.append(1)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_valid, new_y)
```

# Changing the max_depth

```python
train_accuracy = []
validation_accuracy = []
for depth in range(1, 10):
    dt_model = DecisionTreeClassifier(max_depth=depth, random_state=10)
    dt_model.fit(X_train, y_train)
    train_accuracy.append(dt_model.score(X_train, y_train))
    validation_accuracy.append(dt_model.score(X_valid, y_valid))
```

```python
frame = pd.DataFrame({'max_depth':range(1,10), 'train_acc':train_accuracy, 'valid_acc':validation_accuracy})
frame.head()
```

```python
#Using plot we deter mine best value of max_depth. Find all other parameters with for loop
plt.figure(figsize=(12,6))
plt.plot(frame['max_depth'], frame['train_acc'], marker = 'o')
plt.plot(frame['max_depth'], frame['valid_acc'], marker = 'o')
plt.ylabel("Perfomance")
plt.xlabel("Depth of Tree")
plt.legend()
```

# Finding Best max_leaf_nodes

```python
train_accuracy = []
validation_accuracy = []
for leaves in range(2, 30):
    dt_model = DecisionTreeClassifier(max_depth=8, max_leaf_nodes=leaves, random_state=10)
    dt_model.fit(X_train, y_train)
    train_accuracy.append(dt_model.score(X_train, y_train))
    validation_accuracy.append(dt_model.score(X_valid, y_valid))
```

```python
#Using plot we deter mine best value of max_depth. Find all other parameters with for loop
plt.figure(figsize=(12,6))
plt.plot(frame['max_depth'], frame['train_acc'], marker = 'o', label = "Perfomance")
plt.plot(frame['max_depth'], frame['valid_acc'], marker = 'o', label = "Depth of Tree")
plt.ylabel("Perfomance")
plt.xlabel("Depth of Tree")
plt.legend()
plt.show()
```

# Finding best min_sample_leaf

```python

train_accuracy = []
validation_accuracy = []
for sample in range(2, 30):
    dt_model = DecisionTreeClassifier(max_depth=8, max_leaf_nodes=25, min_samples_leaf=sample, random_state=10)
    dt_model.fit(X_train, y_train)
    train_accuracy.append(dt_model.score(X_train, y_train))
    validation_accuracy.append(dt_model.score(X_valid, y_valid))
frame = pd.DataFrame({'max_depth':range(2,30), 'train_acc':train_accuracy, 'valid_acc':validation_accuracy})
frame.head()
#Using plot we deter mine best value of max_depth. Find all other parameters with for loop
plt.figure(figsize=(12,6))
plt.plot(frame['max_depth'], frame['train_acc'], marker = 'o', label = "Perfomance")
plt.plot(frame['max_depth'], frame['valid_acc'], marker = 'o', label = "Depth of Tree")
plt.ylabel("Perfomance")
plt.xlabel("Depth of Tree")
plt.legend()
plt.show()
```
